# Modelo de reinforcement learning no CartPole-v1

## Importando gymnasium(Gym) e Stable Baselines3

```python
import gymnasium as gym
from stable_baselines3 import A2C
```

## Importando o ambiente virtual cartpole-1
- O objetivo do jogo Ã© equilibrar o carrinho usando a alavanca

```python
env = gym.make("CartPole-v1", render_mode="rgb_array")
```

## Criando o agente e treinando por 1.000.000 de vezes

```python
model = A2C("MlpPolicy", env, verbose=1)
```

```python
model.learn(total_timesteps=1_000_000)
```

## Vectorizando o ambiente

```python
vec_env = model.get_env()
```

## Observando o Agente depois de treinar

```python
obs = vec_env.reset()
for i in range(1000):
    action, _state = model.predict(obs, deterministic=True)
    obs, reward, done, info = vec_env.step(action)
    vec_env.render("human")
```
